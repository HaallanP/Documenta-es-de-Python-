**#loc.[ ]** -Selecionando uma linha que corresponde a um rótulo:

```
#Criando DataFrame
df = pd.DataFrame({'A': [-1, 2.6, 3],
              'B': [4, -5, np.nan],
              'C': [0.9, -8, 9],
              'D': [-10, np.nan, 12]
})
#Selecionando uma linha que corresponde a um rótulo:
df_1 = df.loc[1]
df_2 = df.loc[:,['A','D']]
df_3 = df.iloc[1]

display(df_1)
display(df_2)
display(df_3)
```

```
# Filtrando valores menores que 1
resultado = df[df < 1]

print(resultado)
```

```
#filtrando valores, tendo resposta em Booleanos
df_filt = [df["A"] > 1]
df_filt
```

Criando uma tabela de intervalo a partir de um formato de datas

* date_range(), incluindo start, end, freq, e periods

  *O parâmetro **freq** em **pd.date_range()** define a frequência com que as datas serão geradas. Ele controla o intervalo entre as datas. Aqui estão alguns exemplos:

* 'D': Diariamente
* 'W': Semanalmente
* 'M': Mensalmente (no final do mês)
* 'Q': Trimestralmente
* 'H': Horariamente
* 'T' ou 'min': A cada minuto
* 'S': A cada segundo

```
# Criar um intervalo de datas com intervalos de datas
dates = pd.date_range("2013-01-01", periods=6)

# Criar um DataFrame com as datas como uma coluna
df = pd.DataFrame(dates, columns=['Data'])
df['A'] = [1, 2, 3, 4, 5, 6]  # Exemplo de adicionar outras colunas

display(df)
```

##Dados Ausentes  

```
#Criando DataFrame
df2 = pd.DataFrame({'A': [-1, 2.6, 3],
              'B': [4, -5, np.nan],
              'C': [0.9, -8, 9],
              'D': [-10, np.nan, 12]
})

```

```
#Descarta todas as linhas que têm dados ausentes
display(df2)
```

```
#Obtém a máscara booleana onde os valores são:nan
df2 = pd.isna(df2)
df2
```

```
#descartou todas as linhas que tem NaN
print ('\n Descartou linhas que tinha valores NaN')
display(df2.dropna(how="any"))
```

```
#preenchendo dados ausentes
assd = df2.fillna(value= 1000)
print('\n Dados Inseridos a Onde Estava NaN')
display(assd)
```

* ##  **Estatísticas**

```
#Criando DataFrame
df3 = pd.DataFrame({'A': [-1, 2.6, 3],
              'B': [4, -5, np.nan],
              'C': [0.9, -8, 9],
              'D': [-10, np.nan, 12],
              'E': [np.nan, np.nan, 3]
})

```

```
#média
resul = df3.mean()
display(resul)
```

```
#coloccando a media no valores nulos
resul_media = df3.fillna(resul)
display(resul_media)



```

* LAMBDA


```
#Lambda, para cada x(valor) dentro do df e feito a media x 1
Lamb = df3.agg(lambda x: np.mean(x) * 1) #fazendo a média pela linha


display(Lamb)
```

```
#lambda, essa difere da outra, essa pega toda a tabela, multiplicando cada número
lamb2 = resul_media.transform(lambda x: x * 2)# x * 2, fazendo por cada número

display(lamb2)
```

* ## **Agrupamento**
  * Por "agrupar por" estamos nos referindo a um processo envolvendo um ou mais dos Passos a seguir:

  * Dividir os dados em grupos com base em alguns critérios

  * Aplicando uma função a cada grupo de forma independente

  * Combinando os resultados em uma estrutura de dados

```
df4 = pd.DataFrame(
    {
        "A": ["foo", "bar", "foo", "bar", "foo", "bar", "foo", "foo"],
        "B": ["one", "one", "two", "three", "two", "two", "one", "three"],
        "C": np.random.randn(8),
        "D": np.random.randn(8),
    }
)


df4
```

* ## Agrupar por um rótulo de coluna, selecionar rótulos de coluna e, em seguida, aplicar a função ao Grupos:

```
df4.groupby("A")[["C", "D"]].sum()

```

* ## Categóricas
  * Os pandas podem incluir dados categóricos em um arquivo . Para documentos completos, consulte a introdução categórica e a documentação da API.

```
df = pd.DataFrame(
    {"id": [1, 2, 3, 4, 5, 6], "raw_grade": ["a", "b", "b", "a", "a", "e"]}
)
```

```
df["grade"] = df["raw_grade"].astype("category")

df["grade"]
```

```
new_categories = ["very good", "good", "very bad"]
```

```
df["grade"] = df["grade"].cat.rename_categories(new_categories)
```

```
df["grade"] = df["grade"].cat.set_categories(
    ["very bad", "bad", "medium", "good", "very good"])
```

```
df["grade"]
```

```
df.sort_values(by="grade")
```

```
df.groupby("grade", observed=False).size()
```

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Assuming you want a daily time series for 1000 days:
ts = pd.date_range('1/1/2023', periods=1000)

df = pd.DataFrame(
    np.random.randn(1000, 4), index=ts, columns=["A", "B", "C", "D"]
) # Fixed: Removed "E" from columns to match the shape of the random data (1000, 4)


df = df.cumsum()

plt.figure()

df.plot()

plt.legend(loc='best')
plt.show()
```
